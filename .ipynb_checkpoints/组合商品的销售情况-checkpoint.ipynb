```python
import pymysql
import pandas as pd
from sqlalchemy import create_engine
import warnings
warnings.filterwarnings("ignore")
```

```python
query = '''
select 
a.order_id '订单'
,a.operation_param '详细信息'
,a.user_account '客户'
,a.create_time '订单创建时间'
from t_sys_order_logs a 
left join t_order_main b 
on a.order_id = b.order_id
where a.order_status = -1 
and a.operation_param like '%ZH%'  
and b.order_status  <> 5
'''
```

```python
def account(host,user,passwd):
    
    host = host
    port = 3306
    user = user
    passwd = passwd
    db = 'lichi'
    charset = 'utf8'
    con = pymysql.connect(host = host,port = port,user = user,passwd = passwd,db = db,charset = charset)
    
    return con
  
def deal_with(file,area):
    
    file['详细信息'] = file['详细信息'].str.replace(r'productSku|quantity', '', regex=True)
    file['详细信息'] = file['详细信息'].str.replace('},{','|',regex = True)
    file['详细信息'] = file['详细信息'].str.replace(r'[":\[\]{}"]', '', regex=True)
    file['地区'] = area
    
    return file
    
```

```python
con1 = account('rm-uf67u6788ax303x46go.mysql.rds.aliyuncs.com' , 'lichiyy' , 'P%u7*lL@NwPND&hEfu')
con2 = account('rm-2ze2h731i5s62k78d3o.mysql.rds.aliyuncs.com' , 'lichiyy' , 'LT4qkzRFKw63wpc_')
con3 = account('rm-wz9e30vnkm20y11188o.mysql.rds.aliyuncs.com' , 'lichiyy' , 'lichiyunying1!')
```

```python
file_sh = pd.read_sql(query,con1)
file_bj = pd.read_sql(query,con2)
file_gz = pd.read_sql(query,con3)

file_sh_new = deal_with(file_sh , '上海')
file_bj_new = deal_with(file_bj , '北京')
file_gz_new = deal_with(file_gz , '广州')

```

```python
df_file = pd.concat([file_sh_new,file_bj_new,file_gz_new])
```

```python
df  =df_file.copy()
```

```python
df['详细信息'] = df['详细信息'].str.replace(',','@',regex = True)
df_fz = df.copy()
df = df['详细信息'].str.split('|', expand=True)
df = pd.merge(df_fz,df,how='outer',left_index=True,right_index=True)
df.reset_index(inplace=True)
df = pd.melt(df, id_vars=['订单','地区','详细信息','客户','订单创建时间'])
```
