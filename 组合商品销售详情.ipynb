# 三地合并拼接

```python
df4 = pd.concat([df1,df2])
df5 = pd.concat([df4,df3])
df5['详情'] = df5['详情'].str.replace(',','@',regex = True)
df6 = df5.copy() # 备份
df7 = df6['详情'].str.split('|', expand=True)
df8 = pd.merge(df6,df7,how='outer',left_index=True,right_index=True)
df8.reset_index(inplace=True)
df9 = pd.melt(df8, id_vars=['订单号','地区','详情'])
df9 = df9[['订单号','地区','value']]
```

```python
df9.dropna(subset=['value'],
axis=0, # axis=0表示删除行；
how= 'all', # how=any表示若列name、age中，任意一个出现空值，就删掉该行
inplace=True # inplace=True表示在原df上进行修改；
)
df10 = df9['value'].str.split('@', expand=True)
df11 = pd.merge(df9,df10,how='outer',left_index=True,right_index=True)
df11.rename(columns={0:'sku', 1:'数量'}, inplace=True)
```

```python
df12 = df11[['订单号','地区','sku','数量']]
df15 = df12                                         
def new_long(row):
    if row["sku"][1:2]=="H":
        part1 = np.nan
    else:
        part1 = 1
    finall = part1
    return  finall
df15["备注"]=df15.apply(lambda x:new_long(x),axis=1)
df15 = df15[(df15['备注'] != 1)]
df15 = df15[['订单号','地区','sku','数量']]
df15.to_excel('C:/Users/LC-/Desktop/组合商品详情-组合明细.xlsx',sheet_name='组合商品',index=False)
```

```python
df13 = pd.merge(df12,df11,how='left',left_on='sku',right_on='组合商品编码')
df13
```

```python
def new_sort(row):
    if row["sku"][1:2]=="H":
        part1 = 'NaN'
    else:
        part1 = row["sku"]
    finall = part1
    return  finall
df13["new_val"]=df13.apply(lambda x:new_sort(x),axis=1)
df13['sku编码1'] = df13['sku编码'].astype(str)
df13['new_val1'] = df13['new_val'].astype(str)
df13['编码'] = df13['sku编码1'] + df13['new_val1']
df13['编码'] = df13['编码'].str.replace('nan','',regex = True)
df13['编码'] = df13['编码'].str.replace('NaN','',regex = True)
df14 = df13[['订单号','地区','组合商品编码','编码','数量']]
```
